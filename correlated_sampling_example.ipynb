```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples
n_samples = 1000

# Generate correlated samples
X = np.random.normal(size=n_samples)
Y = 0.5 * X + np.sqrt(0.75) * np.random.normal(size=n_samples)

# Plot the scatterplot
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
